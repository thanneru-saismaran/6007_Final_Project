```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import geopandas
import math
import time
from pyspark.sql.functions import col
np.random_state=42
```

```python
from pyspark import SparkContext,SQLContext
sc = SparkContext()
sqlContext = SQLContext(sc)
```

```python
earth = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
```

```python
sample_geo = sqlContext.read.csv('sample_geo.txt',header=True,sep='\t',inferSchema=True)
sample_geo = sample_geo.toPandas()
```

```python
fig, ax = plt.subplots()
fig.set_figheight(15)
fig.set_figwidth(15)
ax.set_aspect('equal')
earth.plot(ax=ax,color='white', edgecolor='black')
sample_geo.plot(kind='scatter',x='Longitude',y='Latitude',ax=ax,color='r',marker='^')
plt.show()
```
