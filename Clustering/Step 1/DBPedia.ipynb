```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import geopandas
```

```python
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
```

```python
lat,lon,name = [],[],[]
with open('lat_longs.txt') as fd:
  for row in fd:
    lat.append(float(row.split()[0]))
    lon.append(float(row.split()[1]))
    name.append(row.split()[2])
```

```python
lat_longs = pd.DataFrame()
lat_longs['lat'] = lat
lat_longs['long'] = lon
lat_longs['name of page'] = name
```

```python
lat_longs.to_csv('lat_longs.csv',index=None)
```

```python
fig, ax = plt.subplots()
fig.set_figheight(15)
fig.set_figwidth(15)
ax.set_aspect('equal')
world.plot(ax=ax,color='white', edgecolor='black')
lat_longs.plot(kind='scatter',x='long',y='lat',ax=ax,color='r',marker='^')
plt.show()
```
